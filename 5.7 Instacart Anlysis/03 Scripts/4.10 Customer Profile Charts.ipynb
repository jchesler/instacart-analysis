### age vs family status

```python
#create a new column for cost of order
df['order_cost'] = df.groupby('order_id')['prices'].transform('sum')
```

```python
#check output
df[['user_id', 'order_id','order_cost']]
```

```python
#create a table that gropus people by marital status and then aggregate age, order cost data, and days since proir order data
df.groupby('marital_status').agg({'age' : ['max', 'min', 'mean','median'], 'order_cost' : ['mean'], 'days_since_prior_order' : ['mean'], 'order_number' : ['mean']})
```

### different customer profiles

```python
#group by customer profiles
df.groupby('customer_profile').agg({'age' : ['max', 'min', 'mean','median'], 'order_cost' : ['mean'], 'days_since_prior_order' : ['mean'], 'order_number' : ['mean']})
```

```python
#habits based on age group
df.groupby('age_group').agg({'order_cost' : ['mean'], 'days_since_prior_order' : ['mean'], 'order_number' : ['mean']})
```

```python
#habits based on income level
df.groupby('income_level').agg({'age' : ['max', 'min', 'mean','median'], 'order_cost' : ['mean'], 'days_since_prior_order' : ['mean'], 'order_number' : ['mean']})
```
