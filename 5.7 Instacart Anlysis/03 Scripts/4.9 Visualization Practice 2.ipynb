### 3. Histogram of hour_of_day column

```python
#create a histogram representing frequency of orders by the hour
hist = df['hour_of_day'].plot.hist(bins = 24)
```

this histogram represents the frequency of orders during each hour of the day. this histogram shows that orders begin to pick up right before noon and are pretty steady until about 3 or 4pm before they start to decrease.

### 4. Bar Chart: Loyalty Flag

```python
#create a bar chart
#represent distrobution of orders amoungst loyalty category
bar = df['loyalty_flag'].value_counts().plot.bar()
```
