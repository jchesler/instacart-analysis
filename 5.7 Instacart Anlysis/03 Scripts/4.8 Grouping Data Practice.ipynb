```python
#import libraries
import pandas as pd
import numpy as np
import os
```

```python
#import orders product merged pickle
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))
```

```python
#create a dataframe of the first 1 million rows
df = ords_prods_merge [:1000000]
```

```python
#check shape of df
df.shape
```

```python
df.head()
```

```python
#group data by product name
df.groupby('product_name')
```

```python
#group data by department_id and calculate the mean of the order_number
df.groupby('department_id').agg({'order_number' : ['mean']})
```

```python
#another way to aggregate data
df.groupby('department_id')['order_number'].mean()
```

```python
#multiple aggregates
df.groupby('department_id').agg({'order_number' : ['mean', 'min', 'max']})
```

```python
#create a new column max orders
#group by user_id
#find max number of orders
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
#check output
ords_prods_merge.head(100)
```

```python
#create flag column to further group customers based on their loyalty
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```
