### 5. Quality and Consistency Checks

```python
#check for missing values
cust.isnull().sum()
```

While there are a significant number of missing values in the first name column, it doesn't impact the data in any way. We shouldn't remove this data based on the fact that first names are missing.

The only way that this would impact the data is if a specific customer had two user_ids, one with their fisrt and last name, and one with only their last name. 

When checking for duplicates we should make sure to exclude first names so that we can see if any of the other data is a complete duplicate (except first name). That might indicate the customer has two accounts.

```python
#check for duplicates
cust_dups = cust[cust.duplicated()]
```

```python
cust_dups
```

Based on this there are no full duplicates. We will run another check to see if there are any duplicates when we exclude first names.
